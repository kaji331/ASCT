```julia
using AutomaticSingleCellToolbox,JLD2,DataFrames
```

```julia
# Data from https://github.com/immunogenomics/harmony/blob/master/doc/Seurat.html
# It can be downloaded by https://app.box.com/s/4bzihajayebqkyrqnd21vw85qipj19af
objs = jldopen("harmony_stim.jld2")["objs"]
# objs is a vector containing two WsObj structs...
```

```julia
# integrated_obj = Harmony!(objs;max_pca=20,pca_number=20,min_cells=5)
@time integrated_obj = Harmony!(objs;min_cells=5)
```

```julia
@time UMAP!(integrated_obj)
DimensionPoints(integrated_obj;dimension_name="umap",group_name="WsObj_batch",colorscheme=:rainbow,point_size=4,alpha=0.4)
```

```julia
# Clustering!(integrated_obj)
@time Clustering!(integrated_obj;resolution=0.5)
```

```julia
DimensionPoints(integrated_obj;dimension_name="umap",split_name="WsObj_batch",colorscheme=:tableau_20,point_size=5,alpha=0.5,width=700)
```

```julia
@time DE!(integrated_obj;method="lr",latent_name="WsObj_batch")
```

```julia
groupby(integrated_obj.meta["clusters_latest_DE"],:group) |> 
    x -> combine(x) do y
        first(y,10)
end
```

```julia
FeatureHeat(integrated_obj;gene_label_size=3)
```

```julia
# Compare cluster 1 to clusters 2/3 only in batch 1.
DE!(integrated_obj;method="wilcox",group_name="clusters_MC_0.5",trt="1",ctl=["2","3"],sub_group=(["WsObj_batch"],["1"]))
```

```julia
groupby(integrated_obj.meta["clusters_MC_0.5_DE"],:group) |> 
    x -> combine(x) do y
        first(y,10)
end
```
